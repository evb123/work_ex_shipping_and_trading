# Sankey Diagram to Visualise Data Quality Issues

This diagram was presented in a sprint review to add transparency to data quality issues around a product being built off of the merging of shipping assurance and operational databases.

```python
# import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[])

flows=[ 30, -2, -4, -9,-11,-2, -2]

orientations=[ 0, 0, 1,1,1,1,-1]

labels = [ 'Data Issues Identified', 'SME Needs Further Investigation', 'Other Use Case','No affect on Processing', 'Resolved in Tech Team','Answered by SME', 'Back to Business']
sankey = Sankey(ax=ax, scale=0.1, offset=0.15, head_angle=170)
sankey.add(flows=flows, labels = labels,
           orientations=orientations,
            pathlengths=[0.25, 0.25, 0.25, 0.25, 0.5,.4,.2],
          trunklength = 2)
diagrams = sankey.finish()
diagrams[0].texts[1].set_color('r')
diagrams[0].texts[1].set_fontweight('bold')
diagrams[0].texts[1].set_fontsize(20)


diagrams[0].texts[0].set_fontweight('bold')
diagrams[0].texts[0].set_fontsize(20)

diagrams[0].texts[2].set_fontweight('bold')
diagrams[0].texts[2].set_fontsize(20)

diagrams[0].texts[3].set_fontweight('bold')
diagrams[0].texts[3].set_fontsize(20)

diagrams[0].texts[4].set_fontweight('bold')
diagrams[0].texts[4].set_fontsize(20)

diagrams[0].texts[5].set_fontweight('bold')
diagrams[0].texts[5].set_fontsize(20)

diagrams[0].texts[6].set_fontweight('bold')
diagrams[0].texts[6].set_fontsize(20)
diagrams[0].texts[6].set_color('r')
fig.suptitle('Data Quality Issue Tracking - CLRNC Use Case 1', fontsize=40)
plt.show()
```
